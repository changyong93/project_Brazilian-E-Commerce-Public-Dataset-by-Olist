###  년도-일자 컬럼 생성

```python
def generate_from_year_to_day(dataframe, column):
    dataframe["year_month"] = dataframe[column].astype(str).map(lambda x : pd.to_datetime(x[:7]))
#     dataframe["year_month"] = pd.to_datetime(dataframe[column].map(lambda x : x.strftime('%Y-%m')))
    dataframe["year"] = dataframe[column].dt.year
    dataframe["month"] = dataframe[column].dt.month
    dataframe["week"] = dataframe[column].dt.week
    dataframe["day"] = dataframe[column].dt.day
    dataframe["weekday"] = dataframe[column].dt.weekday
    return dataframe
```

```python
data1 = data.groupby(["order_id","customer_unique_id"]).payment_value.sum().reset_index()
data2_sum = data1.groupby("customer_unique_id").agg({'order_id' : 'count',
                                                 'payment_value' : 'sum'})
data2_mean = data1.groupby("customer_unique_id").agg({'order_id' : 'count',
                                                 'payment_value' : 'mean'})
data2_median = data1.groupby("customer_unique_id").agg({'order_id' : 'count',
                                                 'payment_value' : 'median'})


data3_sum = pd.cut(data2_sum.payment_value,bins = 100).reset_index()
data3_mean = pd.cut(data2_mean.payment_value,bins = 100).reset_index()
data3_median = pd.cut(data2_median.payment_value,bins= 100).reset_index()

data_f_sum = pd.merge(data2_sum,data3_sum,on = "customer_unique_id")
data_f_mean = pd.merge(data2_mean,data3_mean,on = "customer_unique_id")
data_f_median = pd.merge(data2_median,data3_median,on = "customer_unique_id")

pivot_sum_count = pd.pivot_table(data = data_f_sum, index = "order_id", columns = "payment_value_y",values = "customer_unique_id",aggfunc = "count")
pivot_mean_count = pd.pivot_table(data = data_f_mean, index = "order_id", columns = "payment_value_y",values = "customer_unique_id",aggfunc = "count")
pivot_meidan_count = pd.pivot_table(data = data_f_median, index = "order_id", columns = "payment_value_y",values = "customer_unique_id",aggfunc = "count")
```

```python
fig,ax = plt.subplots(3,1,figsize = (20,15))

sns.heatmap(pivot_sum_count,cmap = "gnuplot",ax = ax[0],annot = True)
sns.heatmap(pivot_mean_count,cmap = "gnuplot",ax = ax[1],annot = True)
sns.heatmap(pivot_meidan_count,cmap = "gnuplot",ax = ax[2],annot = True)

ax[0].set_title("고객(customer_unique_id)당 주문건수 및 매출 sum",fontweight = "bold", fontsize = 15)
ax[1].set_title("고객(customer_unique_id)당 주문건수 및 매출 mean",fontweight = "bold", fontsize = 15)
ax[2].set_title("고객(customer_unique_id)당 주문건수별 매출 median",fontweight = "bold", fontsize = 15)

plt.subplots_adjust(hspace=1)
plt.show()
```
