# translate by kakao
- 제한사항 : 속도는 빠르나 API키를 받아서 사용할 경우 일 5만 단어 / 비회원 일 1만 단어 번역

```python
import requests
def translate_by_kakao(query, src, target):
    url = 'https://dapi.kakao.com/v2/translation/translate'
    header = {"Authorization" : "KakaoAK {카카오 rest API Key를 적어주세요}"}
    data = {"src_lang" : src,
            "target_lang" : target,
            "query" : query}
    
    response = requests.post(url,headers = header, data = data)
    results = response.json()['translated_text'][0][0]
    return results
```

```python
# from selenium import webdriver
class papago_rtt:
    def __init__(self, driver_path="C:/selenium/chromedriver.exe", src_lang='ko', tgt_lang='en'):
        self.driver_path = driver_path
        self.init_trans = f"https://papago.naver.com/?sk={src_lang}&tk={tgt_lang}"
        self.back_trans = f"https://papago.naver.com/?sk={tgt_lang}&tk={src_lang}"
        self.start_browser()
    
    def start_browser(self):
        self.driver = webdriver.Chrome(self.driver_path)
    
    def random_sleep(self):
        time.sleep(random.randint(10,30)/10)
    
    def __call__(self, query):
        output = self.translation(query=query)
        output = self.translation(query=output, back=True)
        return output
    
    def translation(self,query, back=False):
        self.driver.get(self.back_trans)  if back == False else self.driver.get(self.init_trans)
        self.random_sleep()

        #번역
        self.driver.find_element_by_css_selector('textarea#txtSource').send_keys(query)
        self.driver.find_element_by_css_selector('button#btnTranslate').click()
        self.random_sleep()
        
        #번역 결과
        return self.driver.find_element_by_css_selector('div#txtTarget').text


papago_url = 'https://papago.naver.com/'
driver = papago_rtt(src_lang='ko', tgt_lang='en')
```
