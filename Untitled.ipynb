### 중복데이터 확인 함수 생성

```python
def show_overlap_column_ids(dataset,column):
    lists = dataset[f"{column}"].value_counts().reset_index()
    order_ids = lists.loc[lists[f"{column}"]>1,"index"].tolist()
    dataset_overlap = dataset[dataset[f"{column}"].isin(order_ids)]
    print(f"중복 {column} 개수 : {dataset_overlap[column].nunique()}")
    return dataset_overlap
```

### prod_cat_name = olist_products & product_category_name_translation

* 일부 상품중 nan. pc_gamer, portateis~~ 상품군의 경우 영어로된 상품명이 없음
* 상품명이 nan인 데이터는 삭제
* pc_gamer & protateis~는 대체 카테고리 찾아보기

```python
prduct_category_english = product_category_name_translation.product_category_name.tolist()
olist_products.loc[~olist_products.product_category_name.isin(prduct_category_english),"product_category_name"].unique()
```

* portateis_cozinha_e_preparadores_de_alimentos는 portateis_casa_forno_e_cafe로 합치기

```python
datas = olist_products[olist_products.product_category_name.notnull()==True]
datas.loc[datas.product_category_name.str.contains("^portateis"),"product_category_name"].unique()
```

* pc_gamers는 pc로 합치기

```python
datas = olist_products[olist_products.product_category_name.notnull()==True]
datas.loc[datas.product_category_name.str.contains("pc"),"product_category_name"].unique()
```

```python
olist_products["product_category_name"] = olist_products.product_category_name.str.replace("pc_gamer","pcs")
olist_products["product_category_name"] = olist_products.product_category_name.str.replace("portateis_cozinha_e_preparadores_de_alimentos","portateis_casa_forno_e_cafe")
```
